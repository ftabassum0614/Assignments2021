```python


# create a figure and axis
fig, ax = plt.subplots()

# scatter the gdp and year
ax.scatter(df['year'], df['gdp'])

# set a title and labels
ax.set_title('GDP over the Years')
ax.set_xlabel('year')
ax.set_ylabel('gdp')
```

## Visualization

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the population and year
ax.scatter(df['year'], df['population'])

# set a title and labels
ax.set_title('Population over the Years')
ax.set_xlabel('year')
ax.set_ylabel('population')
```

From the graph above, we can see that there is an upward trend of the population of Bangladesh over the years. 

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the export and year
ax.scatter(df['year'], df['export'])

# set a title and labels
ax.set_title('Export over the Years')
ax.set_xlabel('year')
ax.set_ylabel('export')
```
