## 1. Use a generator to create a list of the first 100 prime numbers. Include a paragraph explaining how a generator works.

```python
def getprimes(x):
    primes = []

    for a in range(1,10000):
        isPrime = True
        for b in range(2,a):
            if(a%b==0):
                isPrime = False
        if isPrime:
            primes.append(a)

        if len(primes) == x:
            return primes

primelist = getprimes(100)
print(primelist)
```

```python
print(primelist)
len(primelist)
```

```python
type(primelist)
```

## How a generator works:

A Python generator is a function that produces a sequence of results. Generators use a nested for loop to create an iterated data structure. It works by maintaining its local state, so that the function can resume again exactly where it left off when called subsequent times. Thus, we can think of a generator as something like a powerful iterator.
